```python
# Task 6 Correlation and regression
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

df = pd.read_excel('Major Assignment Superstore Data.xlsx', "Task 1 (Random Sample)")

corr_coef, p_value = pearsonr(df["Sales"], df["Profit"])

# Check p-value

print("p-value:", p_value)
if p_value < 0.05:
    print('Reject the null hypothesis, there is a linear relationship between Sales and Profit')
else:
    print('There is no linear relationship between Sales and Profit')
```

```python
# Another approach to the task
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st

df = pd.read_excel('Major Assignment Superstore Data.xlsx', "Task 1 (Random Sample)")

corr = df[['Sales', 'Profit']].corr()
print (corr)

plt.title('Profit vs Sales Correlation testing', fontsize = 20, color = 'purple')
plt.xlabel('Sales', fontsize = 15, color = 'g')
plt.ylabel('Profit', fontsize = 15, color = 'g')
plt.plot('Sales', 'Profit', data = df, linestyle = 'none', marker = 'o')

model = st.linregress(df['Sales'], df['Profit'])
print(model)
m=model[0]
c = model[1]
x = df ['Sales']
plt.plot(x, m*x+c, color = 'purple')
plt.show()
```

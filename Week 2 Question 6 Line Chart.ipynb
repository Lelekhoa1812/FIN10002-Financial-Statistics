```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("Disease.csv")
plt.title("Cases by day", fontsize = 25, color="green")
plt.xlabel("Day", fontsize = 20, color="brown")
plt.ylabel("Case", fontsize = 20, color="orange")
plt.plot(df['Day'], df['Cases'])
plt.savefig("Line chart Case.png")
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("Disease.csv")
plt.title("Cases by day", fontsize = 25, color="green")
plt.xlabel("Day", fontsize = 20, color="brown")
plt.ylabel("Case", fontsize = 20, color="orange")
plt.bar(df['Day'], df['Cases'])
plt.savefig("Line chart Case.png")
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("Disease.csv")

plt.title("Cases by day", fontsize=25, color="green")

# remove x and y axis labels
plt.axis('off')

plt.pie(df['Cases'], labels=df['Day'], autopct='%1.1f%%')

plt.savefig("Pie chart Case.png")
plt.show()
```
